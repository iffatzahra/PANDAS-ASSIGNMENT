```python
import glob

path = r'./' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

us_census = pd.concat(li, axis=0, ignore_index=True)
us_census
```

```python
us_census.columns
```

```python
us_census.dtypes
```

```python
us_census.head()
```

```python
us_census[['Men','Women']]=us_census.GenderPop.str.split("_",expand=True) 
us_census.head()
```

```python
us_census.Men.replace('(M)','',regex=True,inplace=True)
us_census.Women.replace('(F)','',regex=True,inplace=True)

us_census.head(10)
```

```python
us_census.drop(columns = ['GenderPop'], inplace=True)
```

```python
us_census.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
us_census.drop(columns = ['Women'], inplace=True)
```

```python
us_census['Income'] = us_census.Income.str.strip('$')
```

```python
us_census = us_census.replace('%','', regex=True)
```

```python
us_census.loc[:,'Hispanic':'Income'] = round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
us_census
```

```python
us_census['Men'] = us_census['Men'].astype(int)
```

```python
us_census['Women'] = us_census['TotalPop'] - us_census['Men']
```

```python
import matplotlib.pyplot as plt
plt.scatter(us_census['Women'],us_census['Income'] ) 
plt.xlabel('Salary $', fontsize=18)
plt.ylabel('Women_Population', fontsize=16)
plt.show()
```
